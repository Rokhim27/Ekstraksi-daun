```python
from google.colab import files
uploaded = files.upload()
```

```python
import cv2
import numpy as np
import pandas as pd
import os
```

```python
import os

os.makedirs("dataset_daun", exist_ok=True)

for filename in uploaded.keys():
    with open(os.path.join("dataset_daun", filename), 'wb') as f:
        f.write(uploaded[filename])
```

```python
folder = "dataset_daun"
data = []

for file in os.listdir(folder):
    if file.endswith(".jpg") or file.endswith(".png"):
        path = os.path.join(folder, file)

        # Baca gambar
        img = cv2.imread(path)

        # Resize biar sama ukuran
        img = cv2.resize(img, (256, 256))

        # Konversi ke HSV
        hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

        # Hitung rata-rata tiap channel
        h_mean = np.mean(hsv[:, :, 0])
        s_mean = np.mean(hsv[:, :, 1])
        v_mean = np.mean(hsv[:, :, 2])

        data.append([file, h_mean, s_mean, v_mean])

# Tampilkan hasil
data
```

```python
df = pd.DataFrame(data, columns=["Nama File", "H Mean", "S Mean", "V Mean"])

# Simpan
df.to_csv("fitur_daun.csv", index=False)
df.to_excel("fitur_daun.xlsx", index=False)

print("Berhasil disimpan!")
df
```
